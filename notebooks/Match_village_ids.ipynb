```
# now check if btp sectors are in the gov dataset, excluding values with two sectors
y = []
count = 0
for x in btp_sect:
  if "-" not in x:
    count += 1
    if x not in gov_sect:
      y.append(x)
      #print(x)

print('number of non hyphenated sectors in btp dataset',count)
print('number of unmatched sectors in btp dataset',len(y))
y

# There are 36 sector values in the btp dataset that are not in the gov dataset
# These are of more concern, as we cannot give them ids if they are not in the government dataset
```

```
# find a "similar spelling identifier" function
# fuzzywuzzy - https://www.datacamp.com/community/tutorials/fuzzy-string-python
from fuzzywuzzy import fuzz

# define function
# input = mismatched word, output = og word, score, closest word
def get_similar(word,possibilities):
  score = 0
  match = ""
  for x in possibilities:
    s = fuzz.ratio(word, x)
    if s > score:
      score = s
      match = x

  print(f'{word} - {match}: {score}% similarity')

# run on all unmatched entries
for x in y:
  get_similar(x,gov_sect)
```
