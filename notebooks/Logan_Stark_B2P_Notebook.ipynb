## Engineering new Feature (Distance from nearest hospital (miles))
  Distance in miles is calculated using a python Library called Haversine. Haversine is able to calculate the distance between 2 sets of Latitude and Longtitude. Using Haversine we made a function called hos_dist_calc that iterates through both dataframes and compiles the results into a list.

### Buliding hos_dist_clac

```
# First I make a base result
# I will use this later to guage if the function
# is working properly

haversine((-2.760833, 29.488056),(-2.445100, 28.985800), unit=Unit.MILES)
```

```
def hos_dist_calc(df):
  """ This functions is designed to calculate the smallest distance
  between a bridge and the nearest hospital. The function takes the
  latitude and longitude of a bridge in the dataframe, and using Haversine, calculates
  the distance in miles to every hospital in hospitals. it will append
  every value to a list then return the min. That minimum value is the appended to the 
  list results. Results is returned at the end and can be used to make a new column in a Dataframe """
  results = []
  for i, row in df.iterrows():
    cord_list = []
    cords = (row["Bridge Opportunity: GPS (Latitude)"],row["Bridge Opportunity: GPS (Longitude)"])
    for j, row in hospitals.iterrows():
      hos_cords = (row["Lat"],row["Long"])
      cord_list.append(haversine(cords, hos_cords, unit=Unit.MILES))
    results.append(min(cord_list))
  return results
```

#### Function for extracting "Elevation"
Function Name: elevationcreation

##### Writing and testing function
  


```
# created a small subset of data to work on 

tester_df = df.head(4) 
```

```
def elevationcreation(df):
  """ This function Iterates through every row in the inputed Dataframe. 
  It will save over the variable s with the contents of the row for "Bridge Opportunity: Comments". 
  It then uses RegEx to identify cases of elevation in the text data. If a match is found it is apppended to 
  an empty list. if there is no match "none" is appended to the list. Upon reaching the end of the Dataframe 
  the list is returned and used to create a new column.""" 
  elev_col = []
  for i, row in df.iterrows():
    s = str(row["Bridge Opportunity: Comments"])
    p = re.search('Elevation:?(\d+\d+\d+)',s)
    if p:
      elev_col.append(p.group())
    else:
      elev_col.append(p)
  return elev_col
```

```
# I save the list returned by our function to the variable listus

listus = elevationcreation(tester_df)
```
