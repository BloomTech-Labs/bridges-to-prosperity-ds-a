```python
# STEP 2
def plot_restaurants(restaurants):
    # generate a new map

    folium_map = folium.Map(location=[40.738, -73.98],
                            zoom_start=13,
                            tiles="CartoDB positron",
                            width='50%')

    # for each row in the data, add a cicle marker
    for index, row in restaurants.iterrows():
        # calculate net departures
        # net_departures = (row["departure count"]-row["arrival count"])
        
        # generate the popup message that is shown on click.
        popup_text = "{}<br> Cuisine: {}<br> Critical violations: {}<br> Grade: {}"
        
        popup_text = popup_text.format(row["dba"],
                          row["cuisine description"],
                          row["critical violations"],
                          row["grade"])
        
        #TESTING
        # print(row['camis'])
        # print(row['dba'])
        # print(row['critical violations'])

        # radius of circles
        # radius = net_departures/20
        radius = (row['critical violations']+1)*.4
        color="#E37222" # tangerine
        
        # choose the color of the marker
        # if row["critical violations"]<2:
        #     color="#E37222" # tangerine
            
        #     # color="#007849" # green
        #     # color="#E37222" # tangerine
        # elif row["critical violations"]<4:
        #     color="#FFCE00" # orange
        #     # color="#0375B4" # blue
        #     # color="#FFCE00" # yellow            
        #     # color="#0A8A9F" # teal
        # else:
        #     color="#FF0000" # orange

        #   choose the color of the marker
        if row["grade"]=='A':
            # color="#E37222" # tangerine
            
            color="#007849" # green
            # color="#E37222" # tangerine
        elif row["grade"]=='B':
            color="#FFCE00" # orange
            # color="#0375B4" # blue
            # color="#FFCE00" # yellow            
            # color="#0A8A9F" # teal
        else:
            color="#FF0000" # tangerine
            # color="#FF0000" # orange

        
            
        
        # add marker to the map
        folium.CircleMarker(location=(row["latitude"],
                                      row["longitude"]),
                            radius=radius,
                            color=color,
                            popup=popup_text,
                            fill=True).add_to(folium_map)
    return folium_map
```
